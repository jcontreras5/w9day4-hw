```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb
```

```python
# How many neighborhood groups are available and which shows up the most?

#4 groups,Manhattan has the most
air_bnb.groupby('neighbourhood_group').count()
```

```python
# Are private rooms the most popular in manhattan?
air_bnb.groupby(['neighbourhood_group','room_type']).count()
#No , in Manhattan, Entire Apts are the most popular
```

```python
# Which hosts are the busiest and based on their reviews?
#If we are basing this on number_of_reviews and reviews_per_month  then 219517861	Sonder (NYC)
air_bnb.groupby(['host_id','host_name']).count().sort_values('number_of_reviews',ascending=False).head(10)



```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby(['neighbourhood_group']).mean().sort_values('price',ascending=False)

```

```python
# Which neighbor hood group has the highest total price?
#It is a tie between Manhattan,Brooklyn,Queens at 10,000 
air_bnb.groupby(['neighbourhood_group']).describe()['price']
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.sort_values('price',ascending=False).head(5)[['host_name','price']]
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
zero=air_bnb[air_bnb['availability_365']==0]
zero[zero['number_of_reviews']>100]
```

```python
# What host has the highest total of prices and where are they located?
air_bnb.sort_values('price',ascending=False).head(1)[['host_name','neighbourhood_group','neighbourhood','price']]
```

```python
# When did Danielle from Queens last receive a review?
loc=air_bnb[air_bnb['neighbourhood_group']=='Queens']
loc[loc['host_name']=='Danielle'].sort_values('last_review',ascending=False)
```
